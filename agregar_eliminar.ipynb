## Borrar columnas

Para eliminar una columna uso df.drop('columna', axis = n), n = 0 son filas y n = 1 son columnas

```python
df_desastre = pd.read_csv('dataframes/REPORTE__DE_EVENTOS_POR_DESASTRES_NATURALES_Y_ANTR_PICOS.csv', delimiter = ',')
df_desastre.head(5)
```

```python
df_desastre.drop('Viviendas Averiadas', axis = 1).head(5)
```

df.drop() solo borra el objeto del output, pero no del dataframe. Para esto, debe añadirse inplace = True

```python
df_desastre.head(5)
```

```python
df_desastre.drop('Viviendas Averiadas', axis = 1, inplace = True)
df_desastre.head(5)
```

También puedo borrarlos del dataframe asociándolo directamente al dataframe

```python
df_desastre = df_desastre.drop('Puentes Obstruidos', axis = 1)
df_desastre.head(5)
```

Para borrar filas se usa axis = 0

```python
df_desastre.drop(0, axis=0).head(5)
```

Se pueden borrar rangos con range(start,end)

```python
df_desastre.drop(range(0,10), axis=0, inplace=True)
df_desastre.head(5)
```
