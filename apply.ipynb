```python
import pandas as pd
```

```python
df_desastre = pd.read_csv('dataframes/REPORTE__DE_EVENTOS_POR_DESASTRES_NATURALES_Y_ANTR_PICOS.csv', delimiter = ',')
df_desastre.head(5)
```

## Apply

Se define una función

```python
def text(value):
    return str(value) + ' Familias damnificadas'
```

Se utiliza .apply(función) sobre una columna de nuestro dataframe para aplicar la función a todos sus valores

```python
df_desastre['Familias Damnificadas/ Afectadas'].apply(text)
```

Añadir columna nueva basada en una función

```python
df_desastre['Texto_familias'] = df_desastre['Familias Damnificadas/ Afectadas'].apply(text)
df_desastre.head(5)
```

Se pueden usar funciones de tipo lambda

```python
df_desastre['Vías destruidas * 10'] = df_desastre['Vias Destruidas'].apply(lambda x : x*10)
df_desastre.head(5)
```

Operar con distintas columnas y usando condicionales, se aplica a todo el dataframe

```python
df_desastre['Puentes Obstruidos*3 Mov_masa'] = df_desastre.apply(lambda x : x['Puentes Obstruidos'] * 3 if x['EVENTO'] == 'MOVIMIENTO EN MASA' else x['Puentes Obstruidos'], axis = 1) 
#axis = 1 se usa para aclarar que son columnas

df_desastre.head(5)
```
