# Aplicar pivot table

- index = columna del df que se quiere usar como índice
- columns = columna del df que se quiere usar para definir las columnas
- values = llenar los valores de la tabla
- aggfunc = función de agrupamiento

```python
df_desastre.pivot_table(index='EVENTO', columns='MUNICIPIO', values='Familias Damnificadas/ Afectadas') 
```

Los valores en la siguiente tabla representarán la suma total de familias damnificadas o afectadas para cada evento en cada fecha. Evolución de la suma de familias damnificadas o afectadas a lo largo del tiempo para cada tipo de evento.

```python
df_desastre.pivot_table(index='FECHA', columns='EVENTO', values='Familias Damnificadas/ Afectadas', aggfunc=sum) 
```

## Melt

El método melt toma las columnas del DataFrame y las pasa a filas, con dos nuevas columnas, una para especificar el valor de la antigua columna y otra para el valor que traía:

Dataset antes de melt:

| EVENTO       | MUNICIPIO    |
|--------------|--------------|
| Incendio     | Ciudad A     |
| Inundación   | Ciudad B     |
| Terremoto    | Ciudad C     |

Dataset después de melt:

| variable        | value         |
|-----------------|---------------|
| EVENTO          | Incendio      |
| EVENTO          | Inundación    |
| EVENTO          | Terremoto     |
| MUNICIPIO       | Ciudad A      |
| MUNICIPIO       | Ciudad B      |
| MUNICIPIO       | Ciudad C      |
