## loc e iloc, a diferencia del slicing convencional, SÍ son incluyentes con el start y el end

df.loc[a:b, ['label_1','label_2']] es un slicing más avanzado, muestra todas las filas desde a hasta b y filtra las columnas

```python
df_desastre.loc[1:4, ['MUNICIPIO', 'Fallecidos', 'Familias Damnificadas/ Afectadas']]
```

Puedo afectar los resultados filtrados con df.loc mediante operaciones

```python
df_desastre.loc[:, ['Fallecidos']] + 100
```

```python
df_desastre.loc[:, ['MUNICIPIO']] == 'FLORIDABLANCA'
```

df.iloc es similar a df.loc pero buscando por índice, sin poner el nombre específico del índice

```python
df_desastre.iloc[:, 0:3] #Estoy pidiendo todas las filas y las columnas (índices) de la 0 a la 3
```

```python
df_desastre.iloc[1,3] + ' MODIFICADA'
```
