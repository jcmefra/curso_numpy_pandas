```python
import pandas as pd
import numpy as np
```

## tipos de combinación

- left = prioridad al df de la izquierda
- right = prioridad al df de la derecha
- inner = intersección entre ambos
- outer = todos los valores

## Concat

```python
df1 = pd.DataFrame({'A':['A0', 'A1', 'A2','A3'],
                    'B':['B0', 'B1', 'B2','B3'],
                    'C':['C0', 'C1', 'C2','C3'],
                    'D':['D0', 'D1', 'D2','D3']})


df2 = pd.DataFrame({'A':['A4', 'A5', 'A6','A7'],
                    'B':['B4', 'B5', 'B6','B7'],
                    'C':['C4', 'C5', 'C6','C7'],
                    'D':['D4', 'D5', 'D6','D7']})
```

```python
pd.concat([df1,df2], ignore_index=True) #ignore_index=True reorganiza los índices
```

```python
pd.concat([df1,df2], ignore_index=True, axis = 1) #axis=1 hace la concatenación horizontal
```

## Merge

Merge 1

```python
izq = pd.DataFrame({'key': ['k0', 'k1', 'k2','k3'],
                     'A': ['A0', 'A1', 'A2','A3'],
                     'B': ['B0', 'B1', 'B2','B3']})

der = pd.DataFrame({'key': ['k0', 'k1', 'k2','k3'],
                    'C': ['C0', 'C1', 'C2','C3'],
                    'D': ['D0', 'D1', 'D2','D3']})
```

```python
izq.merge(der, on='key') # merge con elementos en común
```

Merge 2

```python
izq = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
                    'A' : ['A0', 'A1', 'A2','A3'],
                    'B': ['B0', 'B1', 'B2','B3']})

der = pd.DataFrame({'key_2' : ['k0', 'k1', 'k2','k3'],
                    'C' : ['C0', 'C1', 'C2','C3'],
                    'D': ['D0', 'D1', 'D2','D3']})
```

```python
izq.merge(der, left_on='key', right_on='key_2') # merge sin elementos en común
```

Merge 3

```python
izq = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
                    'A' : ['A0', 'A1', 'A2','A3'],
                    'B': ['B0', 'B1', 'B2','B3']})

der = pd.DataFrame({'key_2' : ['k0', 'k1', 'k2',np.nan],
                    'C' : ['C0', 'C1', 'C2','C3'],
                    'D': ['D0', 'D1', 'D2','D3']})
```

```python
der
```

```python
izq.merge(der, left_on='key', right_on='key_2')
```

```python
izq.merge(der, left_on='key', right_on='key_2', how='right') # merge sin elementos en común, how nos indica la prioridad y los valores a traer
```

```python
izq.merge(der, left_on='key', right_on='key_2', how='left')
```

## Join

Join hace un index match, no se hace con columnas específicas

```python
izq_2 = pd.DataFrame({'A': ['A0','A1','A2'],
                    'B':['B0','B1','B2']},
                    index=['k0','k1','k2'])

der_2 =pd.DataFrame({'C': ['C0','C1','C2'],
                   'D':['D0','D1','D2']},
                   index=['k0','k2','k3']) 
```

```python
izq_2.join(der_2)
```

```python
izq_2.join(der_2, how='inner') #igual que en merge, how nos indica cómo hacerla
```

```python
izq_2.join(der_2, how='outer') 
```
