```python
import pandas as pd
```

```python
df_desastre = pd.read_csv('dataframes/REPORTE__DE_EVENTOS_POR_DESASTRES_NATURALES_Y_ANTR_PICOS.csv', delimiter = ',')
df_desastre.head(5)
```

Para agrupar por una columna usamos df.groupy('Columna'), se puede contar, sumar, tomar la media, etc.

```python
df_desastre.groupby('MUNICIPIO').count() # el número de ocurrencias en cada municipio
```

```python
df_desastre.groupby('MUNICIPIO').sum() # sumar los valores de todas las columnas numéricas en cada municipio
```

```python
df_desastre.groupby('MUNICIPIO').sum().loc['BUCARAMANGA'] #Agrupo por municipio en cuanto a suma y escojo una fila correspondiente
```

```python
df_desastre.groupby('MUNICIPIO').sum().reset_index() #Después de realizar la agregación por grupo, esta parte del código reinicia el índice del DataFrame resultante
```

```python
df_desastre.groupby('MUNICIPIO').agg(['min','max']) #funciones de agregación min y max para calcular los valores mínimos y máximos en cada grupo respectivamente
```

```python
df_desastre.groupby('MUNICIPIO').agg({'Familias Damnificadas/ Afectadas': ['min','max']}) #se usa un diccionario para aplicar las funciones min y max a determinada columna
```

Llaves compuestas con combinaciones de datos

```python
df_desastre.groupby(['MUNICIPIO', 'EVENTO']).count() #crear grupos basados en diferentes combinaciones de municipio y tipo de evento presentes en los datos y contar las ocurrencias
```
