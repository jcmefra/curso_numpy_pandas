```python
import pandas as pd
```

```python
df_desastre = pd.read_csv('dataframes/REPORTE__DE_EVENTOS_POR_DESASTRES_NATURALES_Y_ANTR_PICOS.csv', delimiter = ',')
df_desastre.head(2)
```

Se pueden realizar filtros que siguen una lógica booleana

```python
afect_vias = df_desastre['Vias Destruidas'] >= 1 #Establecer condiciones
```

```python
df_desastre[afect_vias] #Llamar df con determinada condición
```

Usar dos condiciones para filtrar

```python
bmanga = df_desastre['MUNICIPIO'] == 'BUCARAMANGA' #definir segunda condición
```

```python
df_desastre[afect_vias & bmanga] #Solo mostrar >= 1 vías destruidas en Bucaramanga
```

Para negar una condición se usa la virguilla ~

```python
df_desastre[~afect_vias].head(5)
```
